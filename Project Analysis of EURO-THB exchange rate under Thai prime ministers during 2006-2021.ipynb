#  Project: Analysis of EURO-THB exchange rate under Thai prime ministers during 2006-2021

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib.style as style
```

```python
exchange_rates = pd.read_csv('euro-daily-hist_1999_2020.csv')
```

```python
exchange_rates.rename(columns={'[Thai baht ]': 'Thai_baht',
                               'Period\\Unit:': 'Time'},
                      inplace=True)
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', inplace=True)
exchange_rates.reset_index(drop=True, inplace=True)
```

```python
euro_to_dollar = exchange_rates[['Time', 'Thai_baht']]
euro_to_dollar['Thai_baht'].value_counts()
```

### Observation

The total number of rows in the file is 5,699, while the Thai_bath column contains only 5,080 rows. It was observed that a certain number of rows are missing, requiring data cleansing prior to analysis.

## Drop all unnecessary rows

```python
euro_to_dollar = euro_to_dollar[euro_to_dollar['Thai_baht'] != '-']
```

## Convert the Thai_baht column to a float

```python
euro_to_dollar['Thai_baht'] = euro_to_dollar['Thai_baht'].astype(float)
```

## Plot euro-dollar exchange rate during 1999-2021

```python
plt.plot(euro_to_dollar['Time'], euro_to_dollar['Thai_baht'])
plt.show()
```

## Plot euro-dollar exchange rate using a moving window of 30 days

```python
euro_to_dollar['rolling_mean'] = euro_to_dollar['Thai_baht'].rolling(30).mean()

plt.plot(euro_to_dollar['Time'], euro_to_dollar['rolling_mean'])
plt.show()
```

## The Four Thai Prime Minister

```python
ts_aps_yl_py = euro_to_dollar.copy()[(euro_to_dollar['Time'].dt.year >= 2001 ) & (euro_to_dollar['Time'].dt.year < 2021)]

taksin = ts_aps_yl_py.copy()[ts_aps_yl_py['Time'].dt.year < 2006]

none = apisit = ts_aps_yl_py.copy()[(ts_aps_yl_py['Time'].dt.year >= 2006) & (ts_aps_yl_py['Time'].dt.year < 2008)]

apisit = ts_aps_yl_py.copy()[(ts_aps_yl_py['Time'].dt.year >= 2008) & (ts_aps_yl_py['Time'].dt.year < 2011)]

yingluk = ts_aps_yl_py.copy()[(ts_aps_yl_py['Time'].dt.year >= 2011) & (ts_aps_yl_py['Time'].dt.year < 2014)]

prayut = ts_aps_yl_py.copy()[(ts_aps_yl_py['Time'].dt.year >= 2014) & (ts_aps_yl_py['Time'].dt.year < 2021)]
```

```python
style.use('fivethirtyeight')

plt.figure(figsize=(24, 10))
ax1 = plt.subplot(2,4,1)
ax2 = plt.subplot(2,4,2)
ax3 = plt.subplot(2,4,3)
ax4 = plt.subplot(2,4,4)
ax5 = plt.subplot(2,1,2)
axes = [ax1, ax2, ax3, ax4, ax5]

for ax in axes:
    ax.set_ylim(28, 60)
    ax.set_yticks([30, 35, 40, 45, 50, 55])
    ax.set_yticklabels(['30','35', '40', '45', '50', '55'], alpha=0.3)
    ax.grid(alpha=0.5)
    
ax1.plot(taksin['Time'], taksin['rolling_mean'], color='#DB4437')
ax1.set_xticklabels(['', '2001', '2002', '2003', '2004', '2005', '2006'], alpha=0.3)
ax1.text(731150.0, 67,'TAKSIN', fontsize=24, weight='bold', color='#DB4437')
ax1.text(731075.0, 63,'(2001-2006)', fontsize=18, weight='bold', alpha=0.3)

ax2.plot(apisit['Time'], apisit['rolling_mean'], color='#4285F4')
ax2.set_xticklabels(['2008', '','', '2009','','', '2010','','', '2011'], alpha=0.3)
ax2.text(733450.0, 67,'APISIT', fontsize=24, weight='bold', color='#4285F4')
ax2.text(733400.0, 63,'(2008-2011)', fontsize=18, weight='bold', alpha=0.3)

ax3.plot(yingluk['Time'], yingluk['rolling_mean'], color='#0F9D58')
ax3.set_xticklabels(['2011', '','', '2012','','', '2013','','', '2014'], alpha=0.3)
ax3.text(734500.0, 67,'YINGLUK', fontsize=24, weight='bold', color='#0F9D58')
ax3.text(734500.0, 63,'(2011-2014)', fontsize=18, weight='bold', alpha=0.3)

ax4.plot(prayut['Time'], prayut['rolling_mean'], color='#F4B400')
ax4.set_xticklabels(['','2014',"", '2016','','2018','', '2020'], alpha=0.3)
ax4.text(736080.0, 67,'PRAYUT', fontsize=24, weight='bold', color='#F4B400')
ax4.text(736020.0, 63,'(2014-2021)', fontsize=18, weight='bold', alpha=0.3)

ax5.plot(taksin['Time'], taksin['rolling_mean'], color='#DB4437')
ax5.plot(none['Time'], none['rolling_mean'], color='#C0C0C0')
ax5.plot(apisit['Time'], apisit['rolling_mean'], color='#4285F4')
ax5.plot(yingluk['Time'], yingluk['rolling_mean'], color='#0F9D58')
ax5.plot(prayut['Time'], prayut['rolling_mean'], color='#F4B400')
ax5.grid(alpha=0.5)
ax5.set_xticks([])

ax1.text(730016.0, 80, "Analysis of the EURO-THB exchange rate during different Thai prime ministers'tenure."
        , fontsize=28, weight='bold')
ax1.text(730016.0, 75, "Study of EURO-THB exchange rate during the administrations of Taksin (2001-2006), Apisit (2008-2011), Yingluk (2011-2014), and Prayut (2014-2021)."
        , fontsize=22)

ax5.text(729800.0, 20, '@CHAKRIT MAFANGNOK' + ' '*290 + 'Source: European Central Bank',
        color='#f0f0f0', backgroundcolor='#4d4d4d', size=14)

plt.show()
```
